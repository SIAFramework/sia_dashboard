```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
import os
```

```python
external_stylesheets = ['htthtml//codepen.io/chriddyp/pen/bWLwgP.css']
```

```python
app = dash.Dash(__name__, external_stylesheets=external_stylesheets)
```

```python
app.layout = html.Div([
    html.Label('Dropdown'),
    dcc.Dropdown(
        options = [
            {'label': 'Amazon', 'value': 'amazon'},
            {'label': 'Facebook', 'value': 'facebook'},
            {'label': 'Twitter', 'value': 'twitter'}
        ],
        value = ['facebook', 'twitter'],
        multi = True,
        placeholder = 'Please select the Platform here',
    ),
    
    html.Label('Radio Items'),
    dcc.RadioItems(
        options = [
            {'label': 'Amazon', 'value': 'amazon'},
            {'label': 'Facebook', 'value': 'facebook'},
            {'label': 'Twitter', 'value': 'twitter'}
        ],
        value = 'twitter'
    ),
    
    html.Label('Checkboxes'),
    dcc.Checklist(
        options = [
            {'label': 'Amazon', 'value': 'amazon'},
            {'label': 'Facebook', 'value': 'facebook'},
            {'label': 'Twitter', 'value': 'twitter'}
        ],
        value = ['facebook', 'twitter']
    ),
    
    html.Label('Text Input'),
    dcc.Input(value = 'twitter', type = 'text', placeholder = 'Please enter the text here'
             ),
    
    html.Label('Text Area'),
    dcc.Textarea(
        placeholder = 'Please enter the text here',
        value = 'This is text area component',
        style = {'width': '100%'}
    ),
    
    html.Label('Slider-1'),
    dcc.Slider(
        min = 0,
        max = 9,
        marks = {i: 'Pages {}'.format(i) if i == 1 else str(i) for i in range(1,10)},
        value = 5,
    ),
    
    html.Label('Slider-2'),
    dcc.Slider(
        min = -5,
        max = 10,
        step = 0.5,
        value = -3
    ),
    
    html.Label('Range Slider-1'),
    dcc.RangeSlider(
        count = 1,
        min = -5,
        max = 10,
        step = 0.5,
        value = [-3,7]
    ),
    
    html.Label('Range Slider-2'),
    dcc.RangeSlider(
        marks = {i: 'Pages {}'.format(i) for i in range(-5,7)},
        min = -5,
        max = 6,
        value = [-3,4]
    )
    
], style = {'columnCount': 1}
)
```

```python
if __name__ == '__main__':
    app.run_server()
```
